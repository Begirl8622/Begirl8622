```python
# import data
import pandas as pd
import numpy as np



df = pd.read_csv('people_datasetnew (1).csv')
df.head()

# convert housing status to lower case for more accurate filtering
df['housing_status'] = df['housing_status'].str.lower().str.strip()

# grouping the data for home loans
potential_homebuyer = df[df['housing_status'] == 'rents'][['Name', 'Email']]
potential_refinance = df[df['housing_status'] == 'own'][['Name', 'Email']]

# grouping the data for potenial checking accounts
age_filter = df[(df['Age'] >=18) & (df['Age'] <=25)][['Name', 'Email']]  # filter for ages 18-25

marital_status_filter = df[df['Marital Status'].isin(['Single', 'Divorced', 'Widowed'])][['Name', 'Email']]   # filter for single persons

# Combining the two filters for potential checking file
potential_checking = potential_checking = df[(df['Age'] >=18) & (df['Age'] <=25) & (df['Marital Status'].isin(['Single', 'Divorced', 'Widowed']))][['Name', 'Email']]

# Extract numerical income from 'Income Range' column
df['Income Range'] = df['Income Range'].str.replace(r'[$,]', '', regex=True)  # remove '$' and ','
df['Income Range'] = df['Income Range'].str.extract(r'(\d+)').astype(float) # extract the first number and convert to float

# define filter_condition before using
filter_condition = df['Income Range'] > 50000

# creating copy to not skew data
filtered_df = df[filter_condition].copy()
# creating filters for rewards credit card
income_filter = df[(df['Income Range']>=60000) & (df['Income Range']<=100000)][['Name', 'Email']]
housing_filter = df[df['housing_status'] == 'own'][['Name', 'Email']]

# Merge and remove duplicates
potential_rewards_creditcard = pd.merge(income_filter, housing_filter, on=['Name', 'Email'], how='inner')
potential_rewards_creditcard.drop_duplicates(inplace=True)

df.reset_index(drop=True)

# personal loan filter
housing_filter_one = df[df['housing_status'].isin(['rents', 'Rent Free'])][['Name', 'Email']]
income_filter_one = df[(df['Income Range']>=20000) & (df['Income Range']<=60000)][['Name', 'Email']]

#combining filters for personal loan
potential_personal_loan = df[(df['housing_status'].isin(['rents', 'Rent Free'])) & (df['Income Range']>=20000) & (df['Income Range']<=60000)][['Name', 'Email']]

# creating a list for investment needs

potential_investment_needs = df['Income Range']>=80000
potential_investment_needs = df[potential_investment_needs][['Name', 'Email']]

# printing a marketing list for all potential products
print("Potential Home Buyers:")
print(potential_homebuyer)
print("\nPotential Refinances:")
print(potential_refinance)
print("\nPotential Checking Accounts:")
print(potential_checking)
print("\nPotential Rewards Credit Cards:")
print(potential_rewards_creditcard)
print("\nPotential Personal Loans:")
print(potential_personal_loan)
print("\nPotential Investment Needs:")
print(potential_investment_needs)



```
